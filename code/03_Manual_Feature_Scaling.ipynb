```python
from sklearn import datasets
import numpy as np
import pandas as pd
```

```python
wine = datasets.load_wine()
X = wine.data
```

```python
X
```

```python
df = pd.DataFrame(X, columns=wine.feature_names)
```

```python
df
```

Normalization

```python
def min_max_normalization(data):
    min_vals = np.min(data, axis=0)
    max_vals = np.max(data, axis=0)
    normalized_data = (data - min_vals) / (max_vals - min_vals)
    return normalized_data
```

```python
X_normalized = min_max_normalization(df)
```

```python
print("Original Data:")
df.head()
```

```python
print("\nNormalized Data:")
X_normalized.head()
```

```python
def standardization_dataframe(dataframe):
    mean_vals = dataframe.mean()
    std_devs = dataframe.std()
    standardized_dataframe = (dataframe - mean_vals) / std_devs
    return standardized_dataframe
```

```python
df_standardized = standardization_dataframe(df)
```
